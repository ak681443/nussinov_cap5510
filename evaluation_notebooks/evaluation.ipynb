```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import Levenshtein as L
%matplotlib inline
```

```python
ground_truth = pd.read_csv('../bin/data/output.txt', header=None)
serial = pd.read_csv('../bin/data/output_serial.txt', header=None)
serial_cache = pd.read_csv('../bin/data/output_serial_cache.txt', header=None)
parallel = pd.read_csv('../bin/data/output_parallel.txt', header=None)
parallel_cache = pd.read_csv('../bin/data/output_parallel_cache.txt', header=None)

ground_truth.columns = ['structure']
serial.columns = ['time_taken', 'structure']
serial_cache.columns = ['time_taken', 'structure']
parallel.columns = ['time_taken', 'structure']
parallel_cache.columns = ['time_taken', 'structure']
```

```python
serial['length'] = serial['structure'].map(lambda x: len(x))
serial_cache['length'] = serial['structure'].map(lambda x: len(x))
parallel['length'] = parallel['structure'].map(lambda x: len(x))
parallel_cache['length'] = parallel['structure'].map(lambda x: len(x))

ground_truth['length'] = ground_truth['structure'].map(lambda x: len(x))
```

```python
def distance(pair):
    return L.distance(pair[0], pair[1])
def aggregate_by_length(distance):
    return pd.concat([ground_truth['length'], distance], axis=1).groupby(['length']).mean()
def evaluate_quality(prediction, ground_truth ):
    result = aggregate_by_length(pd.concat([ground_truth['structure'], prediction['structure']], axis=1).apply(distance, axis=1))
    result = result.reset_index()
    result.columns = ['length', 'distance']
    return result
```

```python
evaluate_quality(serial, ground_truth).plot(x='length', y='distance', kind='scatter')
```

```python
evaluate_quality(serial_cache, ground_truth).plot(x='length', y='distance', kind='scatter')
```

```python
evaluate_quality(parallel, ground_truth).plot(x='length', y='distance', kind='scatter')
```
